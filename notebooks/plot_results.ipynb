```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Increasing rows A, cols B

```python
scale = [1, 2, 5, 10]
time_rowA_colB = [14.88, 34.39, 178.09, 1002.97]
plt.plot(scale, time_rowA_colB, label=f'rows A, cols B')

plt.legend()
plt.xlabel('Scale factor')
plt.ylabel('Time to complete (s)')

plt.savefig('../reports/images/rowsA-colsB.png')
```

Increasing cols A, rows B

```python
scale = [1, 2, 5, 10]
time_colA_rowB = [10.73, 11.01, 10.89, 11.10]
plt.plot(scale, time_colA_rowB, label=f'cols A, rows B')

plt.ylim(5, 20)
plt.legend()
plt.xlabel('Scale factor')
plt.ylabel('Time to complete (s)')

plt.savefig('../reports/images/colsA-rowsB.png')
```

Increasing rows and cols B, rows and cols B

```python
scale = [1, 2, 5, 10]
time_rowcolA_rowcolB = [11.38, 33.49, 197.56, 1314.70]
plt.plot(scale, time_rowcolA_rowcolB, label=f'rows&cols A, rows&cols B')

plt.legend()
plt.xlabel('Scale factor')
plt.ylabel('Time to complete (s)')

plt.savefig('../reports/images/rowscolsA-rowscolsB.png')
```

Increasing block_size parameter

```python
scale = [1, 2, 5, 10]
time_block_size = [1542.22, 236.54, 35.83, 11.63]
plt.plot(scale, time_block_size, label=f'block_size')

plt.legend()
plt.xlabel('Scale factor')
plt.ylabel('Time to complete (s)')

plt.savefig('../reports/images/block-size.png')
```
