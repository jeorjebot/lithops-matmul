```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Increasing rows A, cols B

```python
scale = [1, 2, 5, 10]
time_rowA_colB = [14.88, 34.39, 178.09, 1002.97]
plt.plot(scale, time_rowA_colB, label=f'rows A, cols B')

plt.legend()
plt.xlabel('Scale factor')
plt.ylabel('Time to complete (s)')

plt.savefig('../reports/images/rowsA-colsB.png')
```

Increasing cols A, rows B
